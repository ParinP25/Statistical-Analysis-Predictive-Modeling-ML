# Simple Linear Regression

```python
import pandas as pd
import seaborn as sns

df = pd.read_csv("spotify-2023.csv", encoding="latin-1")
df["streams"] = pd.to_numeric(df["streams"], errors="coerce")
```

```python
df.info()
```

```python
numeric_df = df.select_dtypes(include=['int','float'])
numeric_df.info()
```

Remove the 1 null value in `streams`

```python
numeric_df.dropna(inplace=True)
```

## Predict `streams` using some DataFrame column values.

```python
selected_features = ['in_apple_charts','in_deezer_charts','energy_%']
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

x = numeric_df[selected_features[0]]
y = numeric_df['streams']
```

```python
x.head()
```

```python
y.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.20)
```

```python
lin_reg = LinearRegression()
lin_reg.fit(x_train.to_numpy().reshape(-1, 1),y_train)
```

```python
print('Coefficients:', lin_reg.coef_)
```

```python
print('Intercept', lin_reg.intercept_)
```

Cool. How good is this linear regression model at predicting # streams?

```python
y_pred = lin_reg.predict(x_test.to_numpy().reshape(-1, 1))
```

```python
from sklearn.metrics import mean_squared_error, r2_score

print("MSE: %.2f" % mean_squared_error(y_test, y_pred))
```

```python
print("R2 Score (Coefficient of Determination): %.2f" % r2_score(y_test, y_pred))
```

```python
sns.regplot(data=numeric_df, y='streams',x=selected_features[0])
```

Let's run a linear regression classifier to predict streams on all features in `selected_features`

```python
for f in selected_features:
    # Get X and Y features. X -> Input, Y -> What we want to predict.
    x = numeric_df[f]
    y = numeric_df['streams']

    # Train-Test Split
    x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.20)

    # Create and fit the Linear Regression Model.
    lin_reg = LinearRegression()
    lin_reg.fit(x_train.to_numpy().reshape(-1, 1),y_train)

    # Predict using the 'test' values.
    y_pred = lin_reg.predict(x_test.to_numpy().reshape(-1, 1))

    print(f'Linear Regression: {f} to predict streams')
    print("R2 Score (Coefficient of Determination): %.2f" % r2_score(y_test, y_pred))
    print("MSE: %.2f" % mean_squared_error(y_test, y_pred))
    sns.regplot(data=numeric_df, y='streams',x=f)
    print('-'*80)
    
    
```

# Multi-linear Regression


- Extension of Linear Regression
- One dependent variable (Y) and multiple explanatory variables.
- $Y_i = b_0 + b_1X_1i + b_2X_2i + b_3X_3i + \ldots + b_kX_ki + \varepsilon_i$, where Y is the dependent variable, Xs are the independent variables from 1 to k, and the model is estimated using n observations.




### Assumptions of Multi-linear Regression Model

1. **Linearity:** Existence of a *linear relationship* between the dependent and the explanatory variables.

2. **Independence of Independent Variables:** Observations are independent of each other.
   
3. **Homoskedasticity:** Error term in a regression model is constant. That is, the error term does not vary much as the value of the predictor variable changes.

4. **Independence of Errors:** Errors or residuals from the model are not correlated with each other; they are independent observations.

5. **Normality:** errors or residuals follow a normal distribution, which is important for statistical inference.



```python
import pandas as pd
from sklearn.model_selection import train_test_split
import seaborn as sns
import numpy as np

df = pd.read_csv("spotify-2023.csv", encoding="latin-1")
df["streams"] = pd.to_numeric(df["streams"], errors="coerce")
```

Remove null values!

```python
numeric_df = df.select_dtypes(include=['int','float'])
numeric_df.dropna(inplace=True)
numeric_df.info()
```

##### Data Normalization

```python
sns.histplot(data=df, x="streams")
```

```python
numeric_df["streams"] = np.log(numeric_df["streams"])
sns.histplot(data=numeric_df, x="streams")
```

##### Assumption 1: Linearity

```python
data_to_graph = numeric_df[["streams", "in_spotify_playlists", "in_apple_playlists", "in_deezer_charts", "bpm", "liveness_%"]]
sns.pairplot(data=data_to_graph, diag_kind="hist")
```

Separate the dependent vs independent variables.

```python
x = numeric_df.drop('streams',axis=1)
y = numeric_df['streams']
```

##### Assumption 2:  Independent Variables

Why check for multicolinearity? 
- Multicollinearity leads to **unreliable parameter estimates**, making it challenging to accurately gauge how independent variables impact dependent variables.

Analogy
- Imagine a recipe with various ingredients.
- Each ingredient corresponds to an independent variable in your regression model.
- Picture having two ingredients with very similar tastes.
- When you taste the final dish, you can't distinguish between these similar-tasting ingredients.
- This similarity makes it impossible to identify which ingredient truly contributes to the dish's deliciousness.

Multicollinearity Check: Correlation Matrix

```python
corr = x.corr()

# display the correlation matrix
display(corr)

# plot the correlation heatmap
sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns, cmap='RdBu')
```

Multicollinearity Check: 2. Variance Inflation Factor

- If the VIF for a variable is high, it means that this variable is closely related to one or more of the other variables in the model, and this relationship is causing some uncertainty or "inflation" in the estimates.
- A variance inflation factor exists for each of the predictors in a multiple regression model.

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
import statsmodels.api as sm

# the VIF does expect a constant term in the data, so we need to add one using the add_constant method
vif_data = sm.tools.add_constant(x)

vif_values = pd.Series([variance_inflation_factor(vif_data.values, i) for i in range(vif_data.shape[1])], index=vif_data.columns)
vif_values
```

- A VIF of 1 indicates no multicollinearity.
- A VIF value greater than 1.5 or 2 is considered a sign of moderate multicollinearity, 
- A VIF value of 5 or 10 are often seen as indicating severe multicollinearity.

##### Build the Multilinear Regression Model

```python
# split the data into training and test dataset. 
# 20% test and 80% training dataset
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.20, random_state=34)

x_train = sm.add_constant(x_train)
model = sm.OLS(y_train, x_train)
regress_model = model.fit()
```

Get our model summary 

```python
regress_model.summary()
```

##### Assumption 3: Homoscedasticity
- Homoscedasticity means that the error term does not vary much as the value of the predictor variable changes.
- Not having Homoscedasticity also known as heteroscedasticity means the coefficient estimates are less precise.


Breusch-Pagan is a general test to detect heteroscedasticity. 
  - we won't go into how Breusch-Pagan works but you are free to explore. 
   
  - we will leverage the $stats$ module to get the p-value for the Breusch-Pagan test to determine heteroscedasticity

Hypothesis for the Breusch-Pagan Test
- The null hypothesis for Breusch-Pagan test is that the variances for the errors are equal. 
- The alternate hypothesis (the one youâ€™re testing), is that the variances are not equal.
  
Our goal is to fail to reject the null hypothesis meaning we have no heteroscedasticity.

```python
from statsmodels.stats import diagnostic 

_, pval, __, _ = diagnostic.het_breuschpagan(regress_model.resid, regress_model.model.exog)

if pval > 0.05:
    print("The p-value was {:.4}".format(pval))
    print("We fail to reject the null hypthoesis, so there is no heterosecdasticity.")

else:
    print("The p-value was {:.4}".format(pval))
    print("We reject the null hypthoesis, so there is heterosecdasticity.")
```

```python
x_test = sm.add_constant(x_test)
y_predicted = regress_model.predict(x_test)
y_predicted
```

##### Measure of Error
- Metrics to help quantify how well our model fits the data.
    1. Mean Absolute Error (MAE): average magnitude of the errors between predicted and actual values.
    2. Mean Squared Error (MSE): average of the squared errors, which amplifies the impact of larger errors.
    3. Root Mean Squared Error (RMSE): square root of the MSE and allows for easy interpretation in the same units as the dependent variable.

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
import matplotlib.pyplot as plt


model_mae   = mean_absolute_error(y_test, y_predicted)
model_mse   = mean_squared_error(y_test, y_predicted)
model_rmse  =  np.sqrt(model_mse)

print("MSE {:.5}".format(model_mse))
print("MAE {:.5}".format(model_mae))
print("RMSE {:.5}".format(model_rmse))
```

RMSE of 0.74781 means that On average, the predictions made by the model are off by approximately 0.75 log transformed streams when compared to the actual observed values.

```python
plt.figure(figsize=(8, 8))
plt.scatter(y_predicted, y_test, alpha=0.5)  # alpha controls point transparency
plt.title("Scatterplot of Observed vs. Predicted Values")
plt.xlabel("Predicted Values")
plt.ylabel("Observed Values")
plt.xlim(14, max(y_predicted))   # Adjust the limits 
plt.ylim(14, max(y_test))        # Adjust the limits
plt.plot([14, max(y_predicted)], [14, max(y_predicted)], linestyle='--', color='red') 
plt.grid(True)
```

$R^2$ : tells us how well our data fits the model! 

- Better to look at the adjusted $R^2$ when looking at model summary since this penalizes increased model complexity!

```python
model_r2 = r2_score(y_test, y_predicted)
print("R2: {:.2}".format(model_r2))
```
