Question 4: You will build a logistics regression model to predict whether a song has 500 million streams or not.

```python
# TODO STEP 1:  Create a column called more_than_500_million(binary variable) which indicates whether the number of stream is above 500 million or not (boolean value).

df["more_than_500_million"] = df["streams"] > 500000000
df.head()

```

```python
# TODO STEP 2: remove the streams column
no_streams = df.drop(columns=["streams"])
no_streams.head()
```

```python
# TODO STEP 3: split your dataset into training and test set with a 90-10% split. Train dataset: 90% and test dataset: 10%
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

#numeric_df = df.select_dtypes(include=['int','float'])
#numeric_df.dropna(inplace=True)

selected_features = ['in_spotify_playlists','in_spotify_charts','energy_%']
x = df[selected_features[1]]
y = df['more_than_500_million']

x.head()
y.head()

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.10)
```

```python
# TODO STEP 4: build your logistic regression. You are free to use either statsmodel or sklearn
from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression()
log_reg.fit(x_train.to_numpy().reshape(-1, 1),y_train)
```

```python
# TODO STEP 5: print the model coefficients and the p value. What coefficients are significant and which are not.
from statsmodels.stats.outliers_influence import variance_inflation_factor
import statsmodels.api as sm

from statsmodels.stats import diagnostic

print('Coefficients:', log_reg.coef_)
print('Intercept', log_reg.intercept_)

vif_data = sm.tools.add_constant(x)

x_train = sm.add_constant(x_train)
model = sm.OLS(y_train, x_train)
log_regress_model = model.fit()

_, pval, __, _ = diagnostic.het_breuschpagan(log_regress_model.resid, log_regress_model.model.exog)

if pval > 0.05:
    print("The p-value was {:.4}".format(pval))
else:
    print("The p-value was {:.4}".format(pval))

print("""
The coefficients are not too significant, because they just tell us the variation in the regression.
But the intercept is important, because it gives us an idea of how the graph may be drawn.
      """)
```

```python
# TODO STEP 6: make your prediction on the test dataset
x_test = sm.add_constant(x_test)
y_predicted = log_regress_model.predict(x_test)
y_predicted
```

```python
# TODO STEP 7: choose a metric that you think best will capture your model's performance and apply it on the predicted values.
# TODO Why did you choose the metric?
from sklearn.metrics import mean_squared_error
import numpy as np

model_mse   = mean_squared_error(y_test, y_predicted)
model_rmse  =  np.sqrt(model_mse)

print("RMSE {:.5}".format(model_rmse))
print("""
 I choose this metric, because I thought that it is the most accurate. First, I used mean_squared error to determine
 the impact of the error on the prediction. Then, using MSE, I computed the root mean_squared_error. I think this is
 the most accurate because we can interpret it similarly as with our prediction.
      """)
```
